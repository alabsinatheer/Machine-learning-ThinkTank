```python
from tweepy import Stream
from tweepy.streaming import StreamListener
 
class MyListener(StreamListener):
 
    def on_data(self, data):
        try:
            with open('python.json', 'a') as f:
                f.write(data)
                return True
        except BaseException as e:
            print("Error on_data: %s" % str(e))
        return True
 
    def on_error(self, status):
        print(status)
        return True
 
twitter_stream = Stream(auth, MyListener())
twitter_stream.filter(track=['#python'])
```

```python
# Chap02-03/twitter_client.py 
import os 
import sys 
from tweepy import API 
from tweepy import OAuthHandler 

def get_twitter_auth(): 
  """Setup Twitter authentication. 
 
  Return: tweepy.OAuthHandler object 
  """ 
  try: 
    consumer_key = os.environ['dDvW8kNU4uIpd7ZvdOBNfc0ud'] 
    consumer_secret = os.environ['RWi6n9LFrjnz9pegib7aXUEfhpPKfkNrH0TMBvPsJbYuENObEe'] 
    access_token = os.environ['3051995292-ce5pcQvwCprDUFBtPDy7ajS5EBVUZBpCFUx6qiW'] 
    access_secret = os.environ['TwEJ9nPazWNqk8FNSsLpVf9USsT4BQYi5VHzkBQmRkWQpB'] 
  except KeyError: 
    sys.stderr.write("TWITTER_* environment variables not set\n") 
    sys.exit(1) 
  auth = OAuthHandler(consumer_key, consumer_secret) 
  auth.set_access_token(access_token, access_secret) 
  return auth 

def get_twitter_client(): 
  """Setup Twitter API client. 
 
  Return: tweepy.API object 
  """ 
  auth = get_twitter_auth() 
  client = API(auth) 
  return client 
```
